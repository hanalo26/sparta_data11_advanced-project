```python
# 원본 데이터 기준 id 제외한 모든 컬럼이 동일한 행이 있는지 확인해보기
df = pd.read_csv("DieCasting_Quality_Raw_Data.csv", header=[0,1])

# id 제외한 컬럼으로 중복 확인
check_cols = [col for col in df.columns if col != 'id']

# 중복 여부 확인
is_duplicate = df.duplicated(subset=check_cols, keep=False)

print("=== 중복 행 확인 ===")
print(f"전체 행: {len(df)}개")
print(f"중복 행: {is_duplicate.sum()}개")
print(f"고유 행: {(~is_duplicate).sum()}개")

if is_duplicate.sum() > 0:
    print("\n:warning: 중복 데이터 발견!")
else:
    print("\n:white_check_mark: 중복 없음!")
```
